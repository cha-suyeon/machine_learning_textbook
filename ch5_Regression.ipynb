### 5.1.1 직선 모델
- 직선 방정식 $y(x)=w_0x+w_1$
    - 기울기: $w_0$
    - 절편: $w_1$
    - x에 대한 t의 예측치: $y(x)$

### 5.1.2 제곱 오차 함수
- $J = \frac{1}{N}\Sigma_{n=0}^{N-1} (y_n-t_n)^2 $
- $y_n = y(x) = w_0x+w_1$
- J는 평균제곱오차(mean square error, MSE)로 직선과 데이터 점의 차이 제곱의 평균입니다.
- N으로 나누지 않는 경우인 제곱 합계 오차(sum-of-squares error)를 사용하기도 하지만, 이 책에서는 오차의 크기가 N에 의존하지 않는 평균 제곱 오차를 사용하여 진행

```python
# 리스트 5-1-(6)
from mpl_toolkits.mplot3d import Axes3D

# 평균 오차 함수 -------------------
def mse_line(x, t, w):
    y = w[0] * x + w[1]
    mse = np.mean((y-2)**2)
    return mse

# 계산 -------------------
xn = 100 # 등고선 표시 해상도
w0_range = [-25, 25]
w1_range = [120, 170]

x0 = np.linspace(w0_range[0], w0_range[1], xn)
x1 = np.linspace(w1_range[0], w1_range[1], xn)
xx0, xx1 = np.meshgrid(x0, x1)

J = np.zeros((len(x0), len(x1)))

for i0 in range(xn) :
    for i1 in range(xn) :
        J[i1, i0] = mse_line(X, T, (x0[i0], x1[i1]))

# 표시 -------------------
plt.figure(figsize=(9.5, 4))
plt.subplots_adjust(wspace=0.5)

ax = plt.subplot(1, 2, 1, projection='3d')
ax.plot_surface(xx0, xx1, J, rstride = 10, cstride = 10, alpha = 0.3,
                color = 'blue', edgecolor = 'black')
ax.set_xticks([-20, 0, 20])
ax.set_yticks([120, 140, 160])
ax.view_init(20, -60)

plt.subplot(1, 2, 2)
cont = plt.contour(xx0, xx1, J, 30, colors = 'black',
                   levels = [100, 1000, 10000, 100000])
cont.clabel(fmt = '%1.0f', fontsize = 8)
plt.grid(True)
plt.show()
```

- `np.linspace(start, stop, num, endpoint)`
    - `start(requried)`: 범위의 시작값
    - `stop(requeired)`: 아웃풋의 마지막 값
    - `num(optional)`: 전체 아이템의 개수
    - `endpoint(optional)`: 아웃풋 배열에 stop 값이 포함될 것인지 여부
    
 <br/>
 
 - 직선의 절편을 나타내는 $w_1$는 120cm에서 170cm까지의 50의 범위 설정
 - 직선의 기울기를 나타내는 $w_0$에서도 -25에서 25까지의 50 범위 설정

### 5.1.3 매개 변수 구하기(경사하강법)

```python
# 리스트 5-1-(7)
# 평균 제곱 오차의 기울기 -------------------
def dmse_line(x, t, w):
    y = w[0] * x + w[1]
    d_w0 = 2 * np.mean((y-t)*x)
    d_w1 = 2 * np.mean(y-t)
    return d_w0, d_w1
```

```python
# 리스트 5-1-(8)
d_w = dmse_line(X, T, [10, 165])
print(np.round(d_w,1))
```

```python
# 리스트 5-1-(9)
# 경사 하강법 -------------------
def fit_line_num(x,t):
    w_init = [10.0, 165.0] # 초기 매개 변수
    alpha = 0.001 # 학습률
    i_max = 100000 # 반복의 최대수
    eps = 0.1 # 반복 종류 기울기의 절대값의 한계
    w_i = np.zeros([i_max, 2])
    w_i[0, :] = w_init
    
    for i in range(1, i_max):
        dmse = dmse_line(x, t, w_i[i-1])
        w_i[i, 0] = w_i[i-1, 0] - alpha * dmse[0]
        w_i[i, 1] = w_i[i-1, 1] - alpha * dmse[1]
        if max(np.absolute(dmse)) < eps: # 종료판정, np.absolute는 절대값
            break
    
    w0 = w_i[i, 0]
    w1 = w_i[i, 1]
    w_i = w_i[:i, :]
    return w0, w1, dmse, w_i

# 메인 -------------------
plt.figure(figsize = (4, 4)) # MSE의 등고선 표시

xn = 100 # 등고선 해상도
w0_range = [-25, 25]
w1_range = [120, 170]
x0 = np.linspace(w0_range[0], w0_range[1], xn)
x1 = np.linspace(w1_range[0], w1_range[1], xn)
xx0, xx1 = np.meshgrid(x0, x1)
J = np.zeros((len(x0), len(x1)))

for i0 in range(xn) :
    for i1 in range(xn) :
        J[i1, i0] = mse_line(X, T, (x0[i0], x1[i1]))

cont = plt.contour(xx0, xx1, J, 30, colors = 'black', levels = (100, 1000, 10000, 100000))
cont.clabel(fmt = '%1.0f', fontsize = 8)
plt.grid(True)

# 경사 하강법 호출
W0, W1, dMSE, W_history = fit_line_num(X,T)

# 결과 보기
print('반복 횟수 {0}'.format(W_history.shape[0]))
print('W=[{0:.6f}, {1:.6f}]'.format(W0,W1))
print('dMSE=[{0:.6f}, {1:.6f}]'.format(dMSE[0], dMSE[1]))
print('MSE={0:.6f}'.format(mse_line(X, T, [W0, W1])))
plt.plot(W_history[:, 0], W_history[:,1], '.-',
        color='gray', markersize=10, markeredgecolor='cornflowerblue')
plt.show()
```

- `dmse_line`을 사용한 경사 하강법 `fit_line_num(x, t)`를 [리스트 5-1-(9)]에서 구현함
- fit_line_num(x,t)는 데이터 x,t를 인수로 하여 mse_line을 최소화하는 **w**를 돌려줌
- **w**는 초기 값 `w_init=[10.0, 165.0]`에서 시작하여 `dmse_line`에서 구현한 기울기 **w**를 갱신함
- 갱신 단계의 폭이 되는 학습 비율은 `alpha=0.001`로 함
- **w**이 평평한 곳에 도달하면 갱신을 종료함
- 기울기의 각 요소의 값이 `eps=0.1`보다 작으면 for문에서 빠져 나오게 되어 있음

<br/>

- 처음에는 가파른 계곡으로 진행해 골짜기에 정착하면 계곡의 중앙 부근에 천천히 나아가서 기울기가 거의 없어지는 지점에 도달함

### 5.1.4 선형 모델 매개 변수의 해석해
- 경사 하강법은 반복 계산에 의해 근사값을 구하는 수치 계산법이고, 이러한 풀이를 **수치해**라고 함
- 직선 모델의 경우에서는 근사적인 해석이 아니라 방정식을 해결하여 정확한 해를 구할 수 있음. 이러한 풀이를 **해석해**라고 함
- 해석를 도출하면 문제의 본질을 잘 이해할 수 있고, 다차원 데이터에 대응하며, 곡선의 모델로 확장하기 좋고, 커널법 등의 이해를 도와줌
- 목표: *J가 극소화되는 지점*을 찾기
    - 그 지점의 기울기는 0이다                      

```python
# 리스트 5-1-(11)
# 해석해 -------------------
def fit_line(x,t):
    mx = np.mean(x)
    mt = np.mean(t)
    mtx = np.mean(t*x)
    mxx = np.mean(x*x)
    w0 = (mtx-mt*mx) / (mxx-mx**2)
    w1 = mt - w0 * mx
    return np.array([w0, w1])

# 메인 -------------------
W = fit_line(X, T)
print('w0={0:.3f}, w1={1:.3f}.format(W[0],W[1])')
mse=mse_line(X,T,W)
print('SD={0:.3f} cm'.format(np.sqrt(mse)))
plt.figure(figsize=(4,4))
show_line(W)
plt.plot(X, T, marker='o', linestyle='None',
        color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()
```

### 5.2.3 매개 변수의 해석해

```python
# 리스트 5-1-(16)
# 해석해 -------------------
def fit_plane(x0, x1, t) :
    c_tx0 = np.mean(t * x0) - np.mean(t) * np.mean(x0)
    c_tx1 = np.mean(t * x1) - np.mean(t) * np.mean(x1)
    c_x0x1 = np.mean(x0 * x1) - np.mean(x0) * np.mean(x1)

    v_x0 = np.var(x0)
    v_x1 = np.var(x1)

    w0 = (c_tx1 * c_x0x1 - v_x1 * c_tx0) / (c_x0x1 ** 2 - v_x0 * v_x1)
    w1 = (c_tx0 * c_x0x1 - v_x0 * c_tx1) / (c_x0x1 ** 2 - v_x0 * v_x1)
    w2 = -w0 * np.mean(x0) - w1 * np.mean(x1) + np.mean(t)
    return np.array([w0, w1, w2])
```

```python
# 메인 -------------------
plt.figure(figsize = (6, 5))
ax = plt.subplot(1, 1, 1, projection = '3d')
W = fit_plane(X0, X1, T)

print("w0 = {0:.1f}, w1 = {1:.1f}, w2 = {2:.1f} ".format(W[0], W[1], W[2]))

show_plane(ax, W)
show_data2(ax, X0, X1, T)
mse = mse_plane(X0, X1, T, W)

print("SD = {0:.3f} cm".format(np.sqrt(mse)))
plt.show()
```

## 5.4 선형 기저 함수 모델

```python
# 리스트 5-2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 데이터 로드 -------------------
outfile = np.load('ch5_data.npz')
X = outfile['X']
X_min = outfile['X_min']
X_max = outfile['X_max']
X_n = outfile['X_n']
T = outfile['T']
```

```python
# 리스트 5-2-(2)
# 가우스 함수 -------------------
def gauss(x, mu, s) :
    return np.exp(-(x - mu) ** 2 / (2 * s ** 2))
```

```python
# 리스트 5-2-(3)
# 메인 -------------------
M = 4
plt.figure(figsize = (4, 4))
mu = np.linspace(5, 30, M)
s = mu[1] - mu[0] # (A)
xb = np.linspace(X_min, X_max, 100)

for j in range(M) :
    y = gauss(xb, mu[j], s)
    plt.plot(xb, y, color = 'gray', linewidth = 3)

plt.grid(True)
plt.xlim(X_min, X_max)
plt.ylim(0, 1.2)
plt.show()
```

```python
# 리스트 5-2-(4)
# 선형 기저 함수 모델 -------------------
def gauss_func(w, x) :
    m = len(w) - 1
    mu = np.linspace(5, 30, m)
    s = mu[1] - mu[0]
    y = np.zeros_like(x) # x와 같은 크기로 요소가 0의 행렬 y를 작성

    for j in range(m) :
        y = y + w[j] * gauss(x, mu[j], s)
    y = y + w[m]
    return y
```

```python
# 리스트 5-2-(5)
# 선형 기저 함수 모델 MSE -------------------
def mse_gauss_func(x, t, w) :
    y = gauss_func(w, x)
    mse = np.mean((y - t) ** 2)
    return mse
```

```python
# 리스트 5-2-(6)
# 선형 기저 함수 모델 정확한 솔루션 -------------------
def fit_gauss_func(x, t, m) :
    mu = np.linspace(5, 30, m)
    s = mu[1] - mu[0]
    n = x.shape[0]
    psi = np.ones((n, m + 1))

    for j in range(m) :
        psi[:, j] = gauss(x, mu[j], s)
        psi_T = np.transpose(psi)

    b = np.linalg.inv(psi_T.dot(psi))
    c = b.dot(psi_T)
    w = c.dot(t)
    return w
```

```python
# 리스트 5-2-(7)
# 가우스 기저 함수 표시 -------------------
def show_gauss_func(w) :
    xb = np.linspace(X_min, X_max, 100)
    y = gauss_func(w, xb)
    plt.plot(xb, y, c = [.5, .5, .5], lw = 4)
```

```python
# 메인 -------------------
plt.figure(figsize = (4, 4))
M = 4
W = fit_gauss_func(X, T, M)

show_gauss_func(W)
plt.plot(X, T, marker = 'o', linestyle = 'None', color = 'cornflowerblue', markeredgecolor = 'black')
plt.xlim(X_min, X_max)
plt.grid(True)
mse = mse_gauss_func(X, T, W)

print('W = ' + str(np.round(W, 1)))
print("SD = {0:.2f} cm".format(np.sqrt(mse)))
plt.show()
```

```python
# 리스트 5-2-(12)
# K 분할 교차 검증 -------------------
def kfold_gauss_func(x, t, m, k) :
    n = x.shape[0]
    mse_train = np.zeros(k)
    mse_test = np.zeros(k)

    for i in range(0, k) :
        x_train = x[np.fmod(range(n), k) != i] # (A)
        t_train = t[np.fmod(range(n), k) != i] # (A)
        x_test  = x[np.fmod(range(n), k) == i] # (A)
        t_test  = t[np.fmod(range(n), k) == i] # (A)

        wm = fit_gauss_func(x_train, t_train, m)
        mse_train[i] = mse_gauss_func(x_train, t_train, wm)
        mse_test[i]  = mse_gauss_func(x_test, t_test, wm)
    return mse_train, mse_test
```

## 5.6 새로운 모델의 생성

```python
# 리스트 5-2-(17)
# 모델 A -------------------
def model_A(x, w) :
    y = w[0] - w[1] * np.exp(-w[2] * x)
    return y

# 모델 A 표시 -------------------
def show_model_A(w) :
    xb = np.linspace(X_min, X_max, 100)
    y = model_A(xb, w)
    plt.plot(xb, y, c = [.5, .5, .5], lw = 4)

# 모델 A의 MSE -------------------
def mse_model_A(w, x, t) :
    y = model_A(x, w)
    mse = np.mean((y - t) ** 2)
    return mse
```

```python
# 리스트 5-2-(18)
from scipy.optimize import minimize

# 모델 A의 매개 변수 최적화 -------------------
def fit_model_A(w_init, x, t) :
    res1 = minimize(mse_model_A, w_init, args = (x, t), method = "powell")
    return res1.x
```

```python
# 리스트 5-2-(19)
# 메인 -------------------
plt.figure(figsize = (4, 4))
W_init = [100, 0, 0]
W = fit_model_A(W_init, X, T)

print("w0 = {0:.1f}, w1 = {1:.1f}, w2 = {2:.1f}".format(W[0], W[1], W[2]))
show_model_A(W)

plt.plot(X, T, marker = 'o', linestyle = 'None', color = 'cornflowerblue', markeredgecolor = 'black')
plt.xlim(X_min, X_max)
plt.grid(True)
mse = mse_model_A(W, X, T)

print("SD = {0:.2f} cm".format(np.sqrt(mse)))
plt.show()
```
