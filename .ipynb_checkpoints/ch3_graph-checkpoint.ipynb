#### 3.1.3 3차 함수 $f(x)=(x-2)x(x+2)$ 그리기

```python
# 리스트 2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):
    return (x-2) * x * (x+2)
```

```python
# 리스트 2-(2)
print(f(1))
```

```python
# 리스트 2-(3)
print(f(np.array([1,2,3])))
```

#### 3.1.4 그리는 범위를 결정하기

```python
# 리스트 2-(4)
x = np.arange(-3, 3.5, 0.5)
print(x)
```

- `linspace(n1, n2, n)`하면 범위 n1에서 n2 사이를 일정 간격 n개의 구간으로 나눈 값을 돌려줌

```python
# 리스트 2-(5)
x = np.linspace(-3, 3, 10)
print(np.round(x,2))
```

#### 3.1.6 그래프를 장식하기

```python
# 리스트 2-(7)
# 함수를 정의

def f2(x, w):
    return (x-w) * x * (x+2) # (A) 함수 정의

# x를 정의
x = np.linspace(-3, 3, 100) # (B) x를 100 분할하기

# 차트 묘사
plt.plot(x, f2(x,2), color='black', label='$w=2$') # (C)
plt.plot(x, f2(x,1), color='cornflowerblue',
        label='$w=1$')
plt.legend(loc='upper left') # (E) 범례 표시
plt.ylim(-15, 15) # (F) y축의 범위
plt.title('$f_2(x)$') # (G) 제목
plt.xlabel('$x$') # (H) x 라벨
plt.ylabel('$y$') # (I) y 라벨
plt.grid(True) # (J) 그리드
plt.show()
```

```python
# 리스트 2-(8)
import matplotlib
matplotlib.colors.cnames
```

- color(255,0,0)와 같이 0에서 255까지의 정수 tuple 형으로도 RGB 값 설정 가능
- 범례는 plot의 `label='문자열'`에서 내용을 지정하고 `plt.lengend()`로 표기

#### 3.1.7 그래프를 여러 개 보여주기
- 여러 그래프를 나란히 표시하려면 [리스트 2-(9)]와 같이 `plt.subplot(n1, n2, n)`을 사용하여, 전체를 세로 n1, 가로 n2로 나눈 n번째에 그래프가 그려짐
- `plt.subplot`의 **n**은 특별히 0이 아닌 1부터 시작함

```python
# 리스트 2-(9)
plt.figure(figsize=(10,3)) # (A) figure 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) # (B) 그래프의 간격을 지정
for i in range(6):
    plt.subplot(2,3,i+1) # (C) 그래프 묘사의 위치를 지정
    plt.title(i+1)
    plt.plot(x, f2(x,i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

- `plt.figure(figsize=(w,h))`는 전체 영역의 크기 지정
- 가로 w, 세로 h
- `plt.subplots_adjust(wspace=w, hspace=h)`로 좌우 및 상하 간격 조절

### 3.2 3차원 그래프 그리기
#### 3.2.1 이변수 함수

```python
# 리스트 3-(1)
import numpy as np
import matplotlib.pyplot as plt

# 함수 f3을 의미
def f3(x0, x1):
    r = 2 * x0**2 + x1**2
    ans = r * np.exp(-r)
    return ans

# x0, x1에서 각각 f3을 계산
xn = 9
x0 = np.linspace(-2, 2, xn) # (A)
x1 = np.linspace(-2, 2, xn) # (B)
y = np.zeros((len(x0), len(x1))) # (C)
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1]) # (D)
```

```python
# 리스트 3-(2)
print(x0)
```

- 행렬 y는 print(y)로 표시할 수 있지만 소수 부분이 길게 표시된다.
- numpy 함수 round를 사용하여 `np.round(y,n)`을 활용해 입력 값 y를 소수점 n자리로 반올림

#### 3.2.2 수치를 색으로 표현하기: `pcolor`

```python
# 리스트 3-(4)
plt.figure(figsize=(3.5,3))
plt.gray() # (A)
plt.pcolor(y) # (B)
plt.colorbar() # (C)
plt.show()
```

#### 3.2.3 함수의 표면을 표시: surface

```python
# 리스트 3-(5)
from mpl_toolkits.mplot3d import Axes3D # (A)

xx0, xx1 = np.meshgrid(x0, x1) # (B)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1,1,1,projection='3d') # (C)
ax.plot_surface(xx0, xx1, y, rstride=1,cstride=1, alpha=0.3,
               color='blue', edgecolor='black') # (D)
ax.set_zticks((0, 0.2)) # (E)
ax.view_init(75, -95) # (F)
plt.show()
```

```python
# 리스트 3-(6)
print(x0)
print(x1)
```

```python
# 리스트 3-(7)
print(xx0)
```

```python
# 리스트 3-(8)
print(xx1)
```

- 3차원으로 만들기 위해 subplot 선언시 `projectiob='3d'`로 지정
- `rstride`, `cstride`에 가로 및 세로 몇 개의 선을 긋는지 지정
- alpha는 면의 투명도 지정

#### 3.2.4 등고선으로 표시: counter

```python
# 리스트 3-(9)
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
xx0, xx1 = np.meshgrid(x0, x1) # (A)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # (B)
cont.clabel(fmt='%3.2f', fontsize=8) # (C)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```

- contour 플롯은 해상도를 어느 정도 높이지 않으면 정확하게 표시되지 않는다.
